題目:<br>
讀取STOCK_DAY_0050_202009.csv串聯STOCK_DAY_0050_202010.csv，並且找出open小於close的資料


```python
import pandas as pd
```

## 讀取 STOCK_DAY_0050_202009.csv、STOCK_DAY_0050_202010.csv

```python
df1 = pd.read_csv('STOCK_DAY_0050_202009.csv')
df2 = pd.read_csv('STOCK_DAY_0050_202010.csv')
```

## 串聯

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2], axis=0)
df.reset_index(drop=True, inplace=True)
df
```

## 找出 open 小於 close 的資料

```python
df[df['open'] < df['close']]
```

```python
df[df['open'] < df['close']][['open', 'close']]  # 篩選欄位
```

```python
df.loc[df['open'] < df['close'], ['open', 'close']]  # 篩選欄位
```
