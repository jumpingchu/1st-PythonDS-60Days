```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

---
### Q1: 計算標準常態分配，小於 1 的機率有多大?
![Q1](Q1.png)

```python
## Q1
#  提示: 用 cdf 來算
low = -10
high = 10
x = np.linspace(-10, 10, 20)

cumsum_probs = stats.uniform.cdf(x, low, high-low)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of uniform(-10, 10)')
plt.plot(x, cumsum_probs, "b-", linewidth=2)
plt.show()
```

```python
print(cumsum_probs[10])
```

### Q2: 計算標準常態分配，大於 1，小於 -1 的機率有多大?

### Q3: X~N(2, 4), x 服從常態分配，平均數為 2, 變異數為 4，計算 X 小於 6 的機率有多大?

```python
## Q3
#  提示: 和 Q1, Q2 不同，要先標準化。
mu = 2
sigma = 4
x = np.linspace(-6, 10, 16)
y = stats.norm.pdf(x, mu, sigma)

plt.plot(x, y, "b-", linewidth=2)
plt.grid(True)
plt.title('pdf of Norma(2, 4)')
plt.show()
```

```python
cumsum_probs =  stats.norm.cdf(x, mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of Nomal(0, 1)')
plt.plot(x, cumsum_probs, "b-", linewidth=2)
plt.grid(True)
plt.show()
```
