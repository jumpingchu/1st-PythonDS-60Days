範例目標:<br>
1. 實做繪製圖表
2. 藉由圖表對資料有初步認識，並解釋圖表意義

範例重點:<br>
1. 每一種圖表都有相對應用的情境，分辨情境並繪製對應圖表
2. 圖像化展現，方便對於資料有初步的理解

```python
import pandas as pd
import numpy as np
```

## 單一折線圖

```python
ts = pd.Series(np.random.randn(200), index=pd.date_range('1/1/2020', periods=200))
ts = ts.cumsum()
print(ts)
ts.plot();
```

### 多個折線圖同時畫出

```python
df = pd.DataFrame(np.random.randn(200, 3), index=pd.date_range('1/1/2020', periods=200), columns=["A","B","C"])
df = df.cumsum()
print(df)
df.plot();
```

## 圓餅圖

```python
df = pd.Series(np.random.rand(4), index=["A","B","C","D"], name="title")
print(df)
df.plot.pie();
```

## 長條圖

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.bar();
```

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
df.plot.barh();
```

```python
df.plot.bar(stacked=True);
```

## 箱型圖

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])  # 可使用 list('ABCD')
print(df)
df.boxplot();
```

## 散佈圖

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
print(df)
df.plot.scatter(x='A', y='B');
```

---
## pandas.DataFrame.plot(*args, **kwargs)

### kind: str

`line` : line plot (default)

`bar` : vertical bar plot

`barh` : horizontal bar plot

`hist` : histogram

`box` : boxplot

`kde` : Kernel Density Estimation plot

`density` : same as ‘kde’

`area` : area plot

`pie` : pie plot

`scatter` : scatter plot

`hexbin` : hexbin plot

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=["A","B"])
df.plot.area();
```

```python
df = pd.DataFrame(
        np.random.randint(1, 7, 6000),
        columns = ['one']
)
df['two'] = df['one'] + np.random.randint(1, 7, 6000)
ax = df.plot.hist(bins=12, alpha=0.5)
```

```python
df.plot.density();
```

```python
df.plot.kde();
```

```python
df = pd.DataFrame({'x': np.random.randn(10000),
                   'y': np.random.randn(10000)})
df.plot.hexbin(x='x', y='y', gridsize=20);
```
