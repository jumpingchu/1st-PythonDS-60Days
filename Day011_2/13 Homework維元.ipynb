````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
from IPython.display import display
````

### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
display(df1)
display(df2)
````

````python
# 依照 fruit 欄位做合併
pd.merge(df1, df2, suffixes=('_x', '_y'), on='fruit')
````

````python
# - 依照 index 索引做合併
df1.join(df2)
# ValueError: columns overlap but no suffix specified: Index(['fruit', 'weight', 'price'], dtype='object')
````

### 2. [簡答題] 承上題，請問為什麼依照 'fruit' merge 合併後有些資料不見了？


* 因為 merge 預設採用了 how='inner' 方法合併，只會留下兩邊共同擁有的 fruit 資料

### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

* 因為 join 不會將欄位 overlap，需要額外定義 suffix

````python
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
````
